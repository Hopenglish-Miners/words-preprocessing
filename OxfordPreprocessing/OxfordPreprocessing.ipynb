```python
import pandas as pd
import urllib2
import time



"""
    Examples of Part of HTML we want to find
    <span class="hw" data-headword-id="eat">eat</span>
    <span class="hw" data-headword-id="walk">walk</span>
"""

def regexTextFinder(text):      
  import re
  matches=re.findall(r'data-headword-id=\"(.+?)\"',text)
  return ",".join(matches)

videoDataWordList = pd.read_json("videoData_vs_WordList.json", orient="index")
df = pd.DataFrame(videoDataWordList)
df.to_json('NEW_videoData_vs_WordList.json')

startTime = time.time()
#235 total
for index in range(1,2):
# for index in videoDataWordList:
    startTimeIndex = time.time()
    newWordList = []
    wordsIgnored = []
    for word in videoDataWordList[index]["wordList"]:
#         print "Trying word: ", word
        if word.isalnum():
            response = urllib2.urlopen('https://en.oxforddictionaries.com/definition/'+word)
            html = response.read()
            regexTextFinder(html)
            oxfordWord = regexTextFinder(html).split(',')[0]
            if oxfordWord.isalnum():
                newWordList.append(oxfordWord)
                print "Original word is %s Converted word is %s"% (word, oxfordWord)
            else:
                wordsIgnored.append(word)
                print "Did not find word %s "% word
        else: 
            wordsIgnored.append(word)
            print "Word wasnt alphanumeric ", word
            
    print "Finished index ", index
    finalTimeIndex = time.time()
    print "Finished in time: ", (finalTimeIndex-startTimeIndex)
    
    textFileName = ("newWordsForVideo%d.txt"%index)
    text_file = open(textFileName, "w")
    text_file.write("Index: %d \n" % index)
    videoString = "Video postId: %s \n" % videoDataWordList[index]["postId"]
    text_file.write(videoString)
    text_file.write("New words list:")
    strNewWordList = ' ,'.join(newWordList)
    text_file.write(strNewWordList)
    text_file.close()

#     videoDataWordList[index]["wordList"] = newWordList

df = pd.DataFrame(videoDataWordList)
df.to_json('NEW_videoData_vs_WordList.json')

finalTime = time.time()

duration = finalTime - startTime

print "Running duration ", duration


```
